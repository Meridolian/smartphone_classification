# Split data and targets

```python
from sklearn.model_selection import train_test_split

target_train = df_train["price_range"]
data_train = df_train[["battery_power", "ram", "px_width", "px_height"]]
```

# Now, let's build our models to predict price_range

### import sklearn functions

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score, cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform

models_scores = []
```

### With a simple Linear regression

```python
from sklearn.linear_model import LinearRegression

linearregression_model = make_pipeline(StandardScaler(), LinearRegression())
cv = ShuffleSplit(n_splits=10, test_size=0.25, random_state=0)
scores = cross_val_score(linearregression_model, data_train, target_train, cv=cv)
models_scores.append({"linearregression_model": round((scores.mean() - scores.std()) * 100, 2)})
print(f"Accuracy score by cross-validation combined with hyperparameters "
      f"search:\n{scores.mean():.3f} +/- {scores.std():.3f}")
```

### With a Logistic regression

```python
from sklearn.linear_model import LogisticRegression

pipeline = make_pipeline(StandardScaler(), LogisticRegression())
pipeline.get_params()
```

```python
param_distributions = {
    "logisticregression__C": [1 , 0.5, 0.1, 0.05, 0.01],
    "logisticregression__max_iter": loguniform(50, 500)
}
logisticregression_model = RandomizedSearchCV(pipeline, param_distributions)
cv = ShuffleSplit(n_splits=10, test_size=0.25, random_state=0)
scores = cross_val_score(logisticregression_model, data_train, target_train, cv=cv)
models_scores.append({"logisticregression_model": round((scores.mean() - scores.std()) * 100, 2)})
print(f"Accuracy score by cross-validation combined with hyperparameters "
      f"search:\n{scores.mean():.3f} +/- {scores.std():.3f}")
```

### With KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

pipeline = make_pipeline(StandardScaler(), KNeighborsClassifier())
pipeline.get_params()
```

```python
param_distributions = {
    "kneighborsclassifier__n_neighbors": [3, 5, 7, 10, 15, 20],
    "kneighborsclassifier__weights": ["uniform", "distance"]
}
kneighborsclassifier_model = RandomizedSearchCV(pipeline, param_distributions)
cv = ShuffleSplit(n_splits=10, test_size=0.25, random_state=0)
scores = cross_val_score(kneighborsclassifier_model, data_train, target_train, cv=cv)
models_scores.append({"kneighborsclassifier_model": round((scores.mean() - scores.std()) * 100, 2)})
print(f"Accuracy score by cross-validation combined with hyperparameters "
      f"search:\n{scores.mean():.3f} +/- {scores.std():.3f}")
```

### With SVM

```python
from sklearn.svm import SVC

pipeline = make_pipeline(StandardScaler(), SVC())
pipeline.get_params()
```

```python
param_distributions = {
    "svc__C": loguniform(1, 10)
}
svc_model = RandomizedSearchCV(pipeline, param_distributions)
cv = ShuffleSplit(n_splits=10, test_size=0.25, random_state=0)
scores = cross_val_score(svc_model, data_train, target_train, cv=cv)
models_scores.append({"svc_model": round((scores.mean() - scores.std()) * 100, 2)})
print(f"Accuracy score by cross-validation combined with hyperparameters "
      f"search:\n{scores.mean():.3f} +/- {scores.std():.3f}")
```

### With HistGradientBoostingClassifier

```python
from sklearn.ensemble import HistGradientBoostingClassifier

pipeline = make_pipeline(StandardScaler(), HistGradientBoostingClassifier())
pipeline.get_params()
```

```python
param_distributions = {
    "histgradientboostingclassifier__learning_rate": (0.01, 0.1, 1, 10),
    "histgradientboostingclassifier__max_leaf_nodes": (3, 10, 30)
}
histgradientboostingclassifier_model = RandomizedSearchCV(pipeline, param_distributions)
cv = ShuffleSplit(n_splits=10, test_size=0.25, random_state=0)
scores = cross_val_score(histgradientboostingclassifier_model, data_train, target_train, cv=cv)
models_scores.append({"histgradientboostingclassifier_model": round((scores.mean() - scores.std()) * 100, 2)})
print(f"Accuracy score by cross-validation combined with hyperparameters "
      f"search:\n{scores.mean():.3f} +/- {scores.std():.3f}")
```

### Now let's check all scores
### Each score is calculated by score.mean() - score.std(), to get a better view of scores

```python
best_score = {}
for model_score in models_scores:
    model_name = list(model_score.keys())[0]
    score = model_score[model_name]
    if best_score == {} or score > list(best_score.values())[0]:
        best_score = model_score
    print(f"Model {model_name} : {score}%")
print(f"\nBest score is model {list(best_score.keys())[0]} with {best_score[list(best_score.keys())[0]]}%")
```

### So Logistic regression do the best accuracy with less standard deviation and with its best parameters
