# Import pandas and numpy, read the 2 csv and print them

```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("TP_train.csv")
df_test = pd.read_csv("TP_test.csv", header=None)
```

### Rename df_test columns just like df_train

```python
columns_names = list(df_train.columns)[:-1]
for i in list(df_test.columns):
    df_test.rename(columns={i: columns_names[i]}, inplace=True)
df_test
```

```python
df_train.dtypes
```

```python
df_test.dtypes
```

### As we can see, touch_screen have categorical values, let's transform this to numeric

```python
df_train["touch_screen"].replace({"NO": 0, "YES": 1}, inplace=True)
df_test["touch_screen"].replace({"NO": 0, "YES": 1}, inplace=True)
df_train["touch_screen"]
```

### Let's do the same with targets

```python
df_train["price_range"].replace({
    "LOW": 1,
    "MEDIUM": 2, 
    "HIGH": 3,
    "VERY_HIGH": 4
}, inplace=True)
df_train["price_range"]
```

# Plot every variables to see correlations

```python
import seaborn as sns

corr = df_train.corr()
sns.set(rc = {'figure.figsize':(25, 25)})
correlations = sns.heatmap(corr, cmap="YlGnBu")
correlations.set_title("Correlations between all variables \n", fontsize=30)
```

### As we can see, ram and battery_life influence a lot price_range, and px_widht and px_height influence a little price_range

# Cleaning up this all

### on df_train

```python
# Print NaN values per column
for col in list(df_train.columns):
    print(f"NaN in column {col} : {len(df_train[col].loc[df_train[col].isna()])}")
```

### mobile_wt has no real impact on price_range so replacing it by zeros

```python
df_train["mobile_wt"].fillna(value=0, inplace=True)
len(df_train["mobile_wt"].loc[df_train["mobile_wt"].isna()])
```

### but battery_power has impact on price_range, so let's replace those nan values by the mean of the column

```python
mean = round(df_train["battery_power"].mean(), 0)
df_train["battery_power"].fillna(value=mean, inplace=True)
len(df_train["battery_power"].loc[df_train["battery_power"].isna()])
```

### on df_test

```python
# Print NaN values per column
for col in list(df_test.columns):
    print(f"NaN in column {col} : {len(df_test[col].loc[df_test[col].isna()])}")
```

### Let's clean just like df_train

```python
df_test["mobile_wt"].fillna(value=0, inplace=True)
len(df_test["mobile_wt"].loc[df_test["mobile_wt"].isna()])
```

```python
mean = round(df_test["battery_power"].mean(), 0)
df_test["battery_power"].fillna(value=mean, inplace=True)
len(df_test["battery_power"].loc[df_test["battery_power"].isna()])
```
